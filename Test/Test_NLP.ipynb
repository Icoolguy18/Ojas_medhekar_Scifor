```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


nltk.download('punkt')
nltk.download('stopwords')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)

    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    removed_stopwords = [word for word in word_tokens if word.lower() in stop_words]

    return filtered_text, removed_stopwords


filtered_text, removed_stopwords = remove_stopwords(story)
print(filtered_text)
print(removed_stopwords, '\n',len(removed_stopwords))
```

```python
import string
def remove_punctuation(text):

    translator = str.maketrans('', '', string.punctuation)

    # Use the translation table to remove punctuation from the text
    cleaned_text = text.translate(translator)

    return cleaned_text

cleaned_text = remove_punctuation(story)
print(cleaned_text)
```

```python
def simple_stemming(words):
    rules = {
        "ing": lambda w: w[:-3] if w.endswith("ing") else w,
        "ed": lambda w: w[:-2] if w.endswith("ed") else w,
        "s": lambda w: w[:-1] if w.endswith("s") else w,
    }
    stemmed = []
    for word in words.split():
        stemmed_word = word
        for ending, rule in rules.items():
            if word.endswith(ending):
                stemmed_word = rule(word)
                break
        stemmed.append(stemmed_word)
    return ' '.join(stemmed)



stemmed_story = simple_stemming(story)
print("Original Story:", story)
print("Stemmed Story:", stemmed_story)
```

```python
class SimpleStemmer:
    def __init__(self):
        self.rules = {
            "ing": lambda w: w[:-3] if w.endswith("ing") else w,
            "ed": lambda w: w[:-2] if w.endswith("ed") else w,
            "s": lambda w: w[:-1] if w.endswith("s") else w,
        }

    def stem_word(self, word):
        for ending, rule in self.rules.items():
            if word.endswith(ending):
                return rule(word)
        return word

    def stem_words(self, words):
        stemmed = [self.stem_word(word) for word in words]
        return stemmed

stemmer = SimpleStemmer()
stemmed_words = stemmer.stem_words(words)

print("Original Words:", words)
print("Stemmed Words:", stemmed_words)

```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python

class SimpleLemmatizer:
    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()

    def get_wordnet_pos(self, word):
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV}
        return tag_dict.get(tag, wordnet.NOUN)

    def lemmatize_word(self, word):
        pos = self.get_wordnet_pos(word)
        return self.lemmatizer.lemmatize(word, pos)

    def lemmatize_words(self, words):
        lemmatized = [self.lemmatize_word(word) for word in words]
        return lemmatized


lemmatizer = SimpleLemmatizer()
lemmatized_words = lemmatizer.lemmatize_words(words)

print("Original Words:", words)
print("Lemmatized Words:", lemmatized_words)
```

```python
def calculate_tf(text, target_word):
    words = text.split()

    word_count = words.count(target_word)

    total_words = len(words)

    tf = word_count / total_words if total_words > 0 else 0

    return tf

target_word = "Shivaji"
tf = calculate_tf(story, target_word)

print('Term Frequency (TF) of ',target_word, tf)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

nltk.download('punkt')

def plot_word_frequency(text):
    words = word_tokenize(text)
    words = [word.lower() for word in words if word.isalnum()]
    word_counts = Counter(words)
    common_words = word_counts.most_common(10)

    words, counts = zip(*common_words)

    plt.figure(figsize=(10, 6))
    plt.bar(words, counts, color='blue')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Top 10 Most Common Words')
    plt.xticks(rotation=45)
    plt.show()


plot_word_frequency(story)
```

```python
def categorize_named_entities_from_tokens(tokens):
    text = ' '.join(tokens)

    doc = nlp(text)

    categorized_entities = {
        "PERSON": [],
        "GPE": [],
        "ORG": []
    }

    for ent in doc.ents:
        if ent.label_ in categorized_entities:
            categorized_entities[ent.label_].append(ent.text)

    return categorized_entities


named_entities = categorize_named_entities_from_tokens(words)

print("Categorized Named Entities:")
for category, entities in named_entities.items():
    print(category ,entities)
```
